```python
import random
import time

card = {
    "Diablo": {
        "Health": 100,
        "Attack": 90,
        "Defense": 60
    },
    "Medusa": {
        "Health": 100,
        "Attack": 70,
        "Defense": 70
    },
    "Jester": {
        "Health": 120,
        "Attack": 60,
        "Defense": 90
    },
    "Troll": {
        "Health": 150,
        "Attack": 40,
        "Defense": 94
    },
    "Specter": {
        "Health": 100,
        "Attack": 70,
        "Defense": 70
    },
    "Mist": {
        "Health": 100,
        "Attack": 75,
        "Defense": 65
    },
    "Savage": {
        "Health": 100,
        "Attack": 90,
        "Defense": 50
    },
    "Marauder": {
        "Health": 100,
        "Attack": 85,
        "Defense": 50
    },
    "Wimp": {
        "Health": 110,
        "Attack": 40,
        "Defense": 85
    },
    "Sorcerer": {
        "Health": 100,
        "Attack": 70,
        "Defense": 55
    }
}

# This is the method of printing card information
def print_card(card_name):
    print(f"Name: {card_name}")
    for stat_name, stat_value in card[card_name].items():
        print(f"{stat_name}: {stat_value}")

# This is the method to implement the card battle function
def battle(card1, card2):
    damage = max(0, card[card1]['Attack'] - card[card2]['Defense'])
    card[card2]['Health'] -= damage
    card[card2]['Defense'] = max(0, card[card2]['Defense'] - card[card1]['Attack'])  
    return card[card2]['Health'] > 0  

def game():
    # This stores the number of cards each player will be given in the game. It can be any number, as long as it's less than or qual to the size of the card_deck
    deck_size = 2
    # The following is used to draw randomly n number of cards from the deck, where n = deck_size
    player_deck = random.sample(list(card.keys()), deck_size)
    opponent_deck = random.sample(list(card.keys()), deck_size)

    game_flag = True
    while game_flag and opponent_deck:
        print("Player cards:")
        for card_name in player_deck:
            print()
            print_card(card_name)
        print()

        out_card = random.choice(opponent_deck)
        print("Opponent's card:")
        print_card(out_card)
        print()

        # This is a method to determine whether user input is valid
        battle_card = None
        while battle_card is None:
            print("Choose the card you want to play against:")
            battle_card = input()
            if battle_card not in player_deck:
                print("This card is not in your deck. Choose another one.")
                battle_card = None

        if not battle(battle_card, out_card):
            print(f"{battle_card} has been damaged by {out_card}.")
        else:
            print(f"{out_card} has been defeated by {battle_card}.")
        time.sleep(3)

        if not battle(out_card, battle_card):
            print(f"{out_card} has been damaged by {battle_card}.")
        else:
            print(f"{battle_card} has been defeated by {out_card}.")
        time.sleep(3)

        if card[out_card]['Health'] <= 0:
            opponent_deck.remove(out_card)
        if card[battle_card]['Health'] <= 0:
            player_deck.remove(battle_card)

        print("Next round will start shortly...")
        time.sleep(3)

        if not opponent_deck:
            print("Your opponent has no cards left. You won the game!")
            break

        if not player_deck:
            print("You have no cards left. You lost the game!")
            break

game()
```
